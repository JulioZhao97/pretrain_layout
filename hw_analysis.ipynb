```python
import os
import json
import pandas as pd
import seaborn as sns
from matplotlib import pyplot
```

### scihub

```python
val = json.load(open("/mnt/petrelfs/share_data/zhaozhiyuan/publaynet/layout_scihub/val.json"))
```

```python
hw = []
for image in val["images"]:
    width, height = image["width"], image["height"]
    hw.append([width, height])
```

```python
hw = pd.DataFrame(hw, columns=["width", "height"])
```

```python
sns.scatterplot(data=hw, x="width", y="height")
```
